```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv()
```

```python
engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df_modelo = pd.read_sql("SELECT * FROM raw.modelo", engine)
```

```python
df_modelo.head()
```

```python
pd.read_sql("""
SELECT
    lower(trim(nome)) AS nome,
    marca_id,
    COUNT(*) AS quantidade
FROM raw.modelo
GROUP BY lower(trim(nome)), marca_id
HAVING COUNT(*) > 1;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT cambio
FROM raw.modelo;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT combustivel
FROM raw.modelo;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT cambio
FROM public.stg_modelo;
""", engine)
```

```python
pd.read_sql("""SELECT *
FROM raw.modelo
WHERE nome IS NULL
   OR trim(nome) = '';""", engine)
```
