```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv()
```

```python
engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df_fornecedor = pd.read_sql("SELECT * FROM raw.fornecedor", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT status 
FROM raw.fornecedor
""", engine)
```

```python
pd.read_sql("""
SELECT *
FROM raw.fornecedor
WHERE nullif(
    regexp_replace(telefone, '[^0-9]', '', 'g'),
    ''
) IS NULL;
""", engine)
```

```python
pd.read_sql("""
SELECT
    regexp_replace(telefone, '[^0-9]', '', 'g') AS telefone,
    COUNT(*) AS quantidade
FROM raw.fornecedor
GROUP BY regexp_replace(telefone, '[^0-9]', '', 'g')
HAVING COUNT(*) > 1;
""", engine)
```

```python
pd.read_sql("""
SELECT
    lower(trim(email)) AS email,
    COUNT(*) AS quantidade
FROM raw.fornecedor
GROUP BY lower(trim(email))
HAVING COUNT(*) > 1;
""", engine)
```

```python
pd.read_sql("""
SELECT
    regexp_replace(cnpj, '[^0-9]', '', 'g') AS cnpj,
    COUNT(*) AS quantidade
FROM raw.fornecedor
GROUP BY regexp_replace(cnpj, '[^0-9]', '', 'g')
HAVING COUNT(*) > 1;
""", engine)
```
