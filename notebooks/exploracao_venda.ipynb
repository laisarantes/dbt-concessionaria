```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv()
```

```python
engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df_venda = pd.read_sql("SELECT * FROM raw.venda", engine)
```

```python
df_venda.head()
```

```python
pd.read_sql("""
SELECT DISTINCT status
FROM raw.venda;
""", engine)
```

```python
pd.read_sql("""
SELECT *
FROM raw.venda
WHERE cliente_id IS NULL
   OR funcionario_id IS NULL
   OR loja_id IS NULL
   OR veiculo_id IS NULL
   OR data_venda IS NULL
   OR valor_total IS NULL
   OR status IS NULL;
""", engine)
```

```python
pd.read_sql("""
SELECT venda_id, COUNT(*)
FROM raw.venda
GROUP BY venda_id
HAVING COUNT(*) > 1;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT data_venda
FROM raw.venda
ORDER BY data_venda;
""", engine)
```

```python
pd.read_sql("""
SELECT *
FROM raw.venda
WHERE valor_total <= 0;
""", engine)
```
