```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv()
```

```python
engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df_veiculo = pd.read_sql('SELECT * FROM raw.veiculo', engine)
```

```python
pd.read_sql("""
SELECT DISTINCT cor
FROM raw.veiculo
ORDER BY cor;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT condicao
FROM raw.veiculo;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT status
FROM raw.veiculo;
""", engine)
```

```python
pd.read_sql("""
SELECT placa, COUNT(*) AS quantidade
FROM raw.veiculo
GROUP BY placa
HAVING COUNT(*) > 1;
""", engine)
```

```python
pd.read_sql("""
SELECT chassi, COUNT(*) AS quantidade
FROM raw.veiculo
GROUP BY chassi
HAVING COUNT(*) > 1;
""", engine)
```

```python
pd.read_sql("""
SELECT *
FROM raw.veiculo
WHERE placa IS NULL
   OR chassi IS NULL
   OR preco IS NULL
   OR modelo_id IS NULL
   OR fornecedor_id IS NULL;
""", engine)
```
