```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv()
```

```python
engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df_pagamento = pd.read_sql('SELECT * FROM raw.pagamento', engine)
```

```python
df_pagamento.head()
```

```python
pd.read_sql("""
SELECT DISTINCT forma_pagamento
FROM raw.pagamento;
""", engine)
```

```python
pd.read_sql("""
SELECT DISTINCT status_pagamento
FROM raw.pagamento;
""", engine)
```

```python
pd.read_sql("""
SELECT *
FROM raw.pagamento
WHERE desconto IS NULL
OR data_pagamento IS NULL;
""", engine)
```

```python
pd.read_sql("""
SELECT *
FROM raw.pagamento p
LEFT JOIN raw.venda v
ON p.venda_id = v.venda_id
WHERE v.venda_id IS NULL;
""", engine)
```
