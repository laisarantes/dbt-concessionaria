```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv()
```

```python
engine = create_engine(
    f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df_cliente = pd.read_sql(
    "SELECT * FROM raw.cliente",
    engine
)

df_cliente.head()
```

```python
pd.read_sql("""SELECT DISTINCT nome, cpf, telefone, email, status FROM raw.cliente;""", engine)
```

```python
pd.read_sql(""" SELECT table_name FROM information_schema.tables WHERE table_schema ='raw';""", engine)
```
